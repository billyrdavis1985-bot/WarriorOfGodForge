```python
import json
import os
from collections import Counter
import matplotlib.pyplot as plt
import networkx as nx

# Load all harvests
json_files = [f for f in os.listdir('.') if f.endswith('.json')]
data = []
for file in json_files:
    with open(file, 'r') as f:
        data.append(json.load(f))

# Extract key phrases (simple word count >5 letters, top per node)
def get_key_terms(text, top_n=10):
    words = [w.lower() for w in text.split() if len(w) > 5 and w.isalpha()]
    return [w for w, c in Counter(words).most_common(top_n)]

# Build graph
G = nx.Graph()
nodes = ['gemini', 'claude', 'gpt', 'grok']
for node in nodes:
    G.add_node(node, size=2000)

# Edge weight = shared terms count
for i in range(len(data)):
    terms = {}
    for node in nodes:
        text = data[i].get(node, '')
        terms[node] = set(get_key_terms(text))

    for a in nodes:
        for b in nodes:
            if a < b:
                shared = len(terms[a] & terms[b])
                if shared > 0:
                    if G.has_edge(a, b):
                        G[a][b]['weight'] += shared
                    else:
                        G.add_edge(a, b, weight=shared)

# Draw
pos = nx.spring_layout(G, k=0.5)
weights = [G[u][v]['weight'] * 2 for u,v in G.edges()]
nx.draw(G, pos, with_labels=True, node_color='gold', node_size=3000,
        edge_color='red', width=weights, font_size=12, font_weight='bold')
plt.title("Quad-Braid Tension Graph — Shared Doctrinal Terms (Contradiction Strength)")
plt.show()
```
